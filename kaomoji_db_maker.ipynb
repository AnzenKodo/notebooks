```python
kamoji_txt = open("kaomoji.csv", "w")

with open('data/kaomoji.json', 'r', encoding='utf-8') as f:
  kamoji_json = json.load(f)
  for key in kamoji_json:
    kamoji_json_opts = kamoji_json[key]

    kamoji_txt.write("\"\"\""+key+"\"\"\",\"\"\"")
    for value in kamoji_json_opts["original_tags"]:
        kamoji_txt.write(value+" ")
    for value in kamoji_json_opts["new_tags"]:
        kamoji_txt.write(value+" ")
    kamoji_txt.write("\"\"\"\n")

kamoji_txt.close()
!sed -i '6195d' kaomoji.csv
!sed -i '8851d' kaomoji.csv
!sed -i '11299d' kaomoji.csv

df = pd.read_csv('kaomoji.csv')
df.head()
```

```python
csv_file = open("symbols.csv", "w")

with open('data/symbols.json', 'r', encoding='utf-8') as f:
    json_data = json.load(f)
    for key in json_data:
        # kamoji_json_opts = kamoji_json[key]
        csv_file.write("\"\"\"" + json_data[key]["characters"] + "\"\"\",\"\"\""+  key[1:-1] +"\"\"\"\n")

csv_file.close()
!sed -i '325d' symbols.csv
!sed -i '326d' symbols.csv

df = pd.read_csv('symbols.csv')
df.head()
```
